```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_files = ["a_example",
               "b_little_bit_of_everything.in", 
               "c_many_ingredients.in", 
               'd_many_pizzas.in', 
               'e_many_teams.in']
```

```python
def read_input(file):
    ingredients_count = {}
    pizzas = []
    ingred_num = 0
    with open(file, 'r') as f:
        for i,l in enumerate(f):
            if i==0:
                m, t1, t2, t3 = [int(x) for x in l.split()]
            else:
                num = int(l.split()[0])
                ingredients = [x for x in l.split()[1:]]
                pizzas.append(ingredients)
                for ing in ingredients:
                    if ing not in ingredients_count:
                        ingredients_count[ing] = 0
                        ingred_num += 1
                    ingredients_count[ing] += 1
    return m, t1, t2, t3, ingred_num, ingredients_count, pizzas
```

```python
m, t1, t2, t3, ingred_num, ingred_count, pizzas = read_input(input_files[2])
m, t1, t2, t3, ingred_num
```
